```python
# Importação das bibliotecas necessárias

import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
# Matplotib

# Cria umespaço de eixo, de 0 a 2Pi, com 200 pontos.

x = np.linspace(0,2*np.pi,200)
y= np.sin(x)

# Processamento
fig, ax = plt.subplots()

ax.plot(x,y)

# Saída de dados

plt.show()
```

```python
# Função (x,y) quadratica

def quadratica(x):
    return x*x

x = np.linspace(-4,4,100)
y= quadratica(x)

# Processamento

plt.plot(x,y,'b')
plt.title('Função Quadratica')
plt.xlabel('Eixo x',fontsize=12,color='gray')
plt.ylabel('Eixo y',fontsize=12,color='gray')

# Saída de dados

plt.show()
```

```python
quadratica(50)
```

```python
# Função de primeiro grau

def afim(a,b):
    return a*x+b

x = np.linspace(-4,4,100)
y= afim(1.5,1)

# Processamento

plt.plot(x,y,'g')
plt.title('Função afim')
plt.xlabel('Eixo x',fontsize=12,color='blue')
plt.ylabel('Eixo y',fontsize=12,color='blue')

# Saída de dados

plt.show()
```

```python
# Função exponencial

def exponencial(y):
    return x ** y

x = np.linspace(1,16,100)
y= exponencial(0.5)

# Processamento

plt.plot(x,y,'b')
plt.title('Função Exponencial')
plt.xlabel('Eixo x',fontsize=12,color='blue')
plt.ylabel('Eixo y',fontsize=12,color='blue')

# Saída de dados

plt.show()
```

```python
# Função logaritmica

def logaritmo(y):
    return x ** (1/y)

x = np.linspace(1,16,100)
y= logaritmo(10)

# Processamento

plt.plot(x,y,'r')
plt.title('Função Logaritmica')
plt.xlabel('Eixo x',fontsize=12,color='green')
plt.ylabel('Eixo y',fontsize=14,color='blue'

# Saída de dados

plt.show()
```
